```python
import schemdraw
import schemdraw.elements as elm

# Initialize a drawing
with schemdraw.Drawing() as d:
    d += (v := elm.SourceV().label("E\n80V"))
    d += elm.Line().length(1.5).up()
    d += elm.Resistor().right().label("R1\n300Ω")
    d += elm.Switch(action="open").label("SW\nt=0").dot()
    d.push()

    d += elm.Resistor().label("R2\n500Ω").down()
    d += elm.Inductor().label("L\n2.5mH")
    d += elm.Line().tox(v.start)
    d += elm.Line().to(v.start)
    
    d.pop()
    d += elm.Line().right()
    d += elm.Capacitor().length(6).label("C\n40nF").down()
    d += elm.Line().left().dot()

    d.draw()
```

```python
import schemdraw
import schemdraw.elements as elm

# Create a new drawing
with schemdraw.Drawing() as d:
    # Add a voltage source
    d += (v1 := elm.SourceV().label("10V"))
    d += elm.Line().right().length(1)

    # Add the first resistor (R1)
    d += (r1 := elm.Resistor().right().label("R1\n100Ω"))
    d += elm.Line().right().length(1)

    # Add the second resistor (R2)
    d += (r2 := elm.Resistor().down().label("R2\n200Ω"))
    d += elm.Line().tox(v1.start)

    # Complete the circuit
    d += elm.Line().toy(v1.start)
    d += elm.Line().to(v1.start)

    # Draw the circuit
    d.draw()
```

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    d += (V1 := elm.SourceV().label('10V'))
    d += (R1 := elm.Resistor().right().at(V1.end).label('1kΩ'))
    d.push()
    d += (C1 := elm.Capacitor().down().at(R1.end).label('100nF'))
    d.pop()
    d += (L1 := elm.Inductor().right().at(R1.end).label('1mH'))
    d += elm.Line().down().at(L1.end).to(C1.end)
    d += elm.Line().to(V1.start)
```

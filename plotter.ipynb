```python
import schemdraw
import schemdraw.elements as elm

# Initialize a drawing
with schemdraw.Drawing() as d:
    d += (v := elm.SourceV().label("E"))
    d += elm.Line().length(1.5).up()
    d += elm.Resistor().right().label("R1")
    d += elm.Switch(action="open").label("t = 0").dot()
    d.push()

    d += elm.Resistor().label("R2").down()
    d += elm.Inductor().label("L")
    d += elm.Line().tox(v.start)
    d += elm.Line().to(v.start)
    
    d.pop()
    d += elm.Line().right()
    d += elm.Capacitor().length(6).label("C").down()
    d += elm.Line().left().dot()

    d.draw()
```

```python
import schemdraw
import schemdraw.elements as elm

# Create a new drawing
with schemdraw.Drawing() as d:
    # Add a voltage source
    d += (v1 := elm.SourceV().label("10V"))
    d += elm.Line().right().length(1)

    # Add the first resistor (R1)
    d += (r1 := elm.Resistor().right().label("R1\n100Ω"))
    d += elm.Line().right().length(1)

    # Add the second resistor (R2)
    d += (r2 := elm.Resistor().down().label("R2\n200Ω"))
    d += elm.Line().tox(v1.start)

    # Complete the circuit
    d += elm.Line().toy(v1.start)
    d += elm.Line().to(v1.start)

    # Draw the circuit
    d.draw()
```

```python
import schemdraw
import schemdraw.elements as elm

with schemdraw.Drawing() as d:
    # 電源
    d += (V := elm.SourceV().label("V"))

    # 上方水平線
    d.push()
    d += elm.Line().right()
    d += (R1 := elm.Resistor().label("R1"))
    d += (R2 := elm.Resistor().label("R2"))
    d.pop()

    # 下方水平線
    d.push()
    d += elm.Line().right().at(V.start)
    d += (R3 := elm.Resistor().label("R3"))
    d += (R4 := elm.Resistor().label("R4"))
    d.pop()

    # 垂直連接線
    d += elm.Line().up().at(R3.end)
    d += elm.Line().down().at(R1.end)

    # 對角線上的測量儀器
    d += elm.MeterV().label("Vmeter").at(R1.end).to(R3.end)

    d.draw()
```
